```r
# https://github.com/hadley/rvest
library(dplyr)
library(tidyr)
library(rvest)
library(ggplot2)

tom_sawyer_001 <- read_html("http://www.readcentral.com/chapters/Mark-Twain/The-Adventures-of-Tom-Sawyer-Complete/002")

tom_sawyer_001 <- tom_sawyer_001 %>% html_nodes("p") %>% html_text() %>% paste(collapse = ' ')

tom_sawyer_001
```

```r
last_chap <- 36
caps <- formatC(2:last_chap, width = 3, flag = '0') 
chapters <- rep(0, last_chap-1)

for(i in 1:(last_chap-1)){
    s <- sprintf('http://www.readcentral.com/chapters/Mark-Twain/The-Adventures-of-Tom-Sawyer-Complete/%s', caps[i])
    chapters[i] <- read_html(s) %>% html_nodes("p") %>% html_text() %>% paste(collapse = ' ')
}
```

```r
# http://stackoverflow.com/questions/12427385/how-to-calculate-the-number-of-occurrence-of-a-given-character-in-each-row-of-a
library(stringr)
chapters <- data.frame(Chapters = chapters)
head(chapters)
str_count(chapters[1,], pattern = "Tom")
```

```r
chapters <- chapters %>% mutate(Tom = str_count(Chapters, pattern = "Tom"), 
                              Polly = str_count(Chapters, pattern = "Polly"), 
                              Sid = str_count(Chapters, pattern = "Sid"))

chapters %>% select(Tom, Polly, Sid)
```

```r
chapters <- chapters %>% mutate(Becky = str_count(Chapters, pattern = "Becky"),
                              Huck = str_count(Chapters, pattern = "Huck") + str_count(Chapters, pattern = "Huckleberry"))
chapters %>% select(Tom, Becky, Huck)
```

```r
chapters <- chapters %>% mutate(Robinson = str_count(Chapters, pattern = "Robinson"),
                                Injun = str_count(Chapters, pattern = "Injun"),
                                Muff = str_count(Chapters, pattern = "Muff"))
chapters %>% select(Robinson, Injun, Muff)
```

```r
# http://stackoverflow.com/questions/10212106/creating-grouped-bar-plot-of-multi-column-data-in-r
friendsAndFoes <- chapters %>% select(Tom, Polly, Injun) %>% mutate(chapter = 1:35)
head(friendsAndFoes)

ggplot(friendsAndFoes) + geom_bar(aes(chapter, Tom), stat = "identity", alpha=0.2) + 
    geom_bar(aes(chapter, Polly), stat = "identity") + 
    geom_bar(aes(chapter, Injun), stat = "identity", fill="red", alpha=0.3)
```

```r
friendsAndFoes <- chapters %>% select(Tom, Polly, Injun) %>% mutate(chapter = 1:35)
friendsAndFoes <- friendsAndFoes %>% gather(value, names, 1:3)
head(friendsAndFoes)

ggplot(friendsAndFoes, aes(x=chapter, y=value)) + geom_bar(aes(fill=names), stat = "identity", position="dodge") +
    scale_x_continuous(breaks = 1:35)
```

```r
# Quién es el protagonosta del libro? 
personajes <- chapters %>% 
            mutate_each(funs( . = cumsum(.)), -Chapters) %>% 
            select(10:14)  %>% 
            mutate(chapter = 1:35) %>% 
            gather(names, value, 1:5)

head(personajes)
```

```r
ggplot(personajes) + geom_line(aes(x = chapter, y = value, color = names), size = 2)
```

```r
# En la casa? 
fueraDeCasa <- chapters %>% 
            select(-Becky, -Sid, -Robinson, -Muff) %>% 
            mutate_each(funs( . = cumsum(.)), -Chapters) %>% 
            select(6:9)  %>% 
            mutate(chapter = 1:35) %>% 
            gather(names, value, 1:4)

head(fueraDeCasa)
```

```r
ggplot(fueraDeCasa) + geom_line(aes(x = chapter, y = value, color = names), size = 2)
```

```r
# Ladrones? 
Ladrones <- chapters %>% select(-Becky, -Sid, Polly) %>% mutate_each(funs( . = cumsum(.)), -Chapters) %>% 
        select(11:13)  %>% mutate(chapter = 1:35) %>% gather(value, names, 1:3)

names(Ladrones) <- c("Chapter", "Names", "Count")

ggplot(Ladrones) + geom_line(aes(x = Chapter, y = Count, color = Names), size = 2)
```
