```r
clima <- read.csv('../tablas/bike data/201508_weather_data.csv', 
                  stringsAsFactors=FALSE)
head(clima)
names(clima)
```

```r
clima.bicis <- clima %>% select(PDT, Max.TemperatureF) %>% 
mutate(Max.TemperatureF = round((Max.TemperatureF - 32)*5/9, 0)) %>% 
rename(Fecha = PDT, Max.Temp = Max.TemperatureF) 

head(clima.bicis)
```

```r
dim(trips.bici)
dim(clima.bicis)
```

Problemas con las dimensiones. Voy a agrupar por fecha y tomar la temperatura promedio cen vez de max.temp. 

```r
clima.bicis <- clima.bicis %>% group_by(Fecha) %>% 
    summarise(MM.Temp = mean(Max.Temp))

dim(clima.bicis)
head(clima.bicis)
```

```r
date_temp <- left_join(trips.bici, clima.bicis, by = 'Fecha')  

dim(date_temp)
head(date_temp)
```

## Factores

factores tipo de dato. Es un vector de números enteros con etiquetas. 

c(1, 0, 0, 1) y 1 <- 'mujer', 0 <- 'hombre'

Etiquetas son fáciles de leer. Se llaman labels. Los valores se llaman levels. 

Pueden agrupar factores parecidos definiedo los niveles:

```r
fact <- factor(clima$Events)
levels(fact) <- c('Seco', 'Humedo', 'Humedo', 'Mojado', 'Mojado')
```

```r
levels(fact)
fact[1:100]
```

```r
clima.simple <- clima %>% rename(Fecha = PDT) %>% 
select(Fecha, Events) %>% mutate(Events = fact)

head(clima.simple)
```

```r
clima.simple %>% group_by(Events) %>% summarize(Count = n())
```

```r
clima.94107 <- clima %>% rename(Fecha = PDT, Max.Temp = Max.TemperatureF) %>%  
    mutate(Events = fact)  %>% 
    filter(Zip == 94107) %>% 
    select(Fecha, Events, Max.Temp) 

head(clima.94107)
```

```r
clima.94107 <-  clima.94107 %>% 
mutate(Max.Temp = factor(round((clima.94107$Max.Temp - 32)*5/9, 1) > 15, 
                         labels = c('Frio', 'Caliente')))

dim(clima.94107)
```

```r
dim(trips.bici)

bicis.94107 <- left_join(clima.94107, trips.bici, by = 'Fecha')
head(bicis.94107)
```

```r
bicis.94107 %>% group_by(Max.Temp) %>% 
    summarize(Viajes.Promedio = round(mean(Viajes), 0))
```

```r
clima.viajes <- bicis.94107 %>% group_by(Max.Temp, Events) %>% 
                    summarize(Viajes = round(mean(Viajes), 0))

head(clima.viajes)
```
