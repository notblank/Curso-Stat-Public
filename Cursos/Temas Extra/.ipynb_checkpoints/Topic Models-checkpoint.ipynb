## Text Analysis

```r
user_reviews %>% 
    ggplot(aes(grade)) +
        geom_histogram()
```

```r
library(tidytext)
library(lubridate)
```

```r
user_reviews %>% 
    group_by(week = floor_date(date, "week")) %>% 
    summarise(nb_reviews = n(), 
              grade = mean(grade)) %>% 
    filter(nb_reviews >= 20) %>% 
    ggplot(aes(week, grade)) +
        geom_line() +
        geom_point(aes(size = nb_reviews)) +
        expand_limits(y = 0)
    
```

```r
user_reviw_words <- user_reviews %>% 
                        unnest_tokens(output = word, input = text) %>% 
                        anti_join(stop_words, by = "word") %>% 
                        count(user_name, date, grade, word, sort = TRUE)

head(user_reviw_words)
```

```r
user_reviw_words %>% 
    group_by(word) %>% 
    summarise(nb_reviews = n(), 
              avg_grade = mean(grade)) %>% 
    filter(nb_reviews >= 50) %>% 
    arrange(desc(avg_grade))
```
