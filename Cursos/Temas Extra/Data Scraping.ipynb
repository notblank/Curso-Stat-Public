```r
#https://github.com/rfordatascience/tidytuesday/tree/master/data/2019/2019-06-04
#https://www.youtube.com/watch?v=tCa2di7aEP4
library(tidyverse)
```

```r
ramen_ratings <- readr::read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-06-04/ramen_ratings.csv")
head(ramen_ratings)
```

```r
ramen_rating_processed <- ramen_ratings %>% 
                                    mutate(style = fct_lump(style, 4), 
                                           country = fct_lump(country, 12), 
                                           brand = fct_lump(brand, 20)) %>%
                                    replace_na(list(style = "Other")) %>%
                                    mutate(style = fct_relevel(style, "Pack"),
                                           country = fct_relevel(country, "Other"), 
                                           brand = fct_relevel(brand, "Other"))

head(ramen_rating_processed)
```

```r
ramen_rating_processed %>%
                        gather(category, value, -review_number, -stars) %>%
                        count(category, value) %>%
                        group_by(category) %>%
                        top_n(20) %>%
                        ungroup() %>%
                        mutate(value = drlib::reorder_within(value, n, category)) %>%
                        ggplot(aes(value, n)) + 
                        geom_col() +
                        facet_wrap(~ category, scales = "free_y") + 
                        drlib::scale_x_reordered() +
                        coord_flip() + 
                        labs(title = "Categorical Predictors (After Processing)", 
                             x = "Predictor", y = "")
```

```r
library(broom)
lm(stars ~ country + style + brand,ramen_rating_processed) %>% 
        tidy(conf.int = TRUE) %>% 
        mutate(term = fct_reorder(term, estimate)) %>%
        ggplot(aes(estimate, term)) + 
            geom_point() +
            geom_errorbar(aes(xmin = conf.low, xmax = conf.high))
```

```r
lm(stars ~ country + style + brand,ramen_rating_processed) %>% 
        tidy(conf.int = TRUE) %>% 
        mutate(term = fct_reorder(term, estimate)) %>%
        filter(term != "(Intercept)") %>%
        ggplot(aes(estimate, term)) + 
            geom_point() +
            geom_errorbar(aes(xmin = conf.low, xmax = conf.high)) +
            labs(title = "Coefficients that predict ramen ratings", 
                 subtitle = "Less common brands, countries and styles were used as the refference level", 
                 x = "Estimated effect on ramen ratings", y = "")
```

```r
lm(stars ~ country + style + brand,ramen_rating_processed) %>% 
        tidy(conf.int = TRUE) %>% 
        filter(term != "(Intercept)") %>%
        extract(term, c("category", "term"), "^([a-z]+)([A-Z].*)") %>%
        mutate(term = fct_reorder(term, estimate)) %>%
        ggplot(aes(estimate, term, color = category)) + 
            geom_point() +
            geom_vline(xintercept = 0, lty = 2) +
            geom_errorbar(aes(xmin = conf.low, xmax = conf.high)) +
            labs(title = "Coefficients that predict ramen ratings", 
                 subtitle = "Less common brands, countries and styles were used as the refference level", 
                 x = "Estimated effect on ramen ratings", y = "") +
            facet_wrap(~category, ncol = 1, scales = "free_y") +
            theme(legend.position = "none")
```

```r
library(tidytext)

ramen_rating_processed %>% filter(!is.na(stars)) %>%
                            unnest_tokens(word, variety) %>%
                            group_by(word) %>%
                            summarise(avg_rating = mean(stars), 
                                      n = n()) %>%
                            arrange(desc(n)) %>% head()

#maybe lasso?
```

## Written reviews

```r
review_links <- read_html("https://www.theramenrater.com/resources-2/the-list/") %>%
                        html_nodes("#myTable , a")

review_links <- review_links[133:length(review_links) - 5]
```

```r
reviews <- tibble(review_number = parse_number(html_text(review_links)), 
                       link = html_attr(review_links, "href")) 
```

```r
get_review_text <- function(url){
    message(url)
    
    read_html(url) %>% html_nodes(".entry-content > p") %>%
                        html_text() %>%
                        str_subset(".")
}
```

```r
library(tidytext)
```

```r
review_paragraphs <- review_texts %>% filter(str_detect(text, "stars")) %>%
                                    mutate(text = str_remove(text, "Finished.*?\\."), 
                                           text = str_remove(text, "EAN.*"),
                                           text = str_remove(text, "JAN.*")) 
```

```r
review_paragraphs_tokenized <- review_paragraphs %>% unnest_tokens(word, text) %>%
                                                        filter(str_detect(word, "[a-z]")) %>%
                                                        anti_join(stop_words, by = "word") 
```

```r
review_words <- review_paragraphs_tokenized %>% 
                                inner_join(ramen_rating_processed, by = "review_number") %>% 
                                filter(!is.na(stars)) %>%
                                group_by(word) %>%
                                summarise(n = n(), 
                                          reviews = n_distinct(review_number),
                                          avg_rating = mean(stars)) %>%
                                arrange(desc(reviews))


review_words %>% head()
```

### Corrolation graphs

```r
# pairs of words
library(widyr)

review_words_filtered <- review_words %>% filter(reviews < 200, reviews >= 10) 

word_cors <- review_paragraphs_tokenized %>%
                            filter(!(word %in% c('bar', 'code')), !str_detect(word, ".’")) %>%
                            distinct(review_number, word) %>%
                            semi_join(review_words_filtered, by = "word") %>%
                            pairwise_cor(word, review_number, sort = TRUE)
```

```r
word_cors %>% head()
```

```r
#https://www.tidytextmining.com/ngrams.html
library(ggraph)
```

```r
set.seed(2019)

word_cors_filtered <- word_cors %>% head(300) 
nodes <- review_words_filtered %>% filter(word %in% word_cors_filtered$item1 | word %in% word_cors_filtered$item2)
```

```r
word_cors_filtered %>% igraph::graph_from_data_frame(vertices = nodes) %>%
                                ggraph(layout = 'fr') +
                                geom_edge_link() +
                                geom_node_point(aes(size = reviews, color = avg_rating)) + 
                                geom_node_text(aes(label = name), repel = TRUE) +
                                scale_color_gradient2(low = 'red', high = 'blue', midpoint = 4.0) +
                                labs(color = 'Average rating', size = 'Number of reviews')
```
