Vamos a usar muestreo con probabilidades.

Para eso tenemos que saber la probabilidad de seleccionar cualquier fila (cualquier subconjunto) de antemano. 
Necesitamos un plan para predecir lo que va a pasar. 

Un ejemplo: 

    Tengo 3 colores: R, V y A. Quiero escoger 2 siguiedo algunas reglas: 
        
        1* Siempre eligo el A primero.
        2* Luego eligo R o V con la misma proba. 
        
        
        P(A R) = 1/2
        P(A V) = 1/2
        
     Pero tambi√©n necesito describir la proba de escoger cualquiera de los otros subconjuntos:
     
     P(V A) = 0
     P(A) = 1
     P(V) = P(R) = 1/2
     P(A V R) = 0

```r
randIndex <- sample(x = 1:100, size = 25, replace = TRUE)

sort(randIndex)

movies %>% slice(sort(randIndex))
```

```r
startVal <- sample(x = 1:10, size = 1)
startVal
index <- seq(from = startVal, to = 50000, by = 10)
index

movies %>% slice(index)

# 1/10 proba de tener '68. No es uniforme: algunas combinaciones son imposibles.
# Es un muestreo aleatorio: sabemos la proba de seleccionar cualquier subconjunto.
```
