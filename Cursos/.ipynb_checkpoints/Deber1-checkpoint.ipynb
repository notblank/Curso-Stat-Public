```r
flights %>% filter(!is.na(dep_delay)) %>%
            group_by(year, month, day, hour) %>% 
            summarize(delay=mean(dep_delay), num_vuelos=n()) %>%
            filter(num_vuelos > 10)

# Una tabla que contiene el retaso promedio en cada da y para cada hora. 
# Solo tiene las horas con mas de 10 vuelos. 
# El !is.na() sirve para no tener problemas con la variable delay.  
```

```r
vuelos_diarios <- flights %>% group_by(day, month) %>% 
            summarize(n = n())

vuelos_diarios
mean(vuelos_diarios$n)
```
